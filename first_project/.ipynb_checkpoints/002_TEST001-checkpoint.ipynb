```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import plotly.graph_objects as go
import datetime
import warnings

mpl.rcParams['axes.unicode_minus'] = False
plt.rc('font', family='NanumBarunGothic')
warnings.filterwarnings('ignore')

# df_covid 전처리 (2019-01-01 ~ 2021-07-31)
df_covid = pd.read_csv('./DATA/002_2.COVID-19.csv', encoding='utf-8', parse_dates=['DATE'])


# df_covid['DATE']= df_covid['DATE'].astype('str')
df_covid['DATE'] = pd.to_datetime(df_covid['DATE'].astype('str'))

df_covid.rename(columns= {'ID' : 'COUNT'}, inplace=True)
df_covid = df_covid.groupby(['DATE']).count()
df_covid.insert(0, 'date', df_covid.index)

df_covid = df_covid[{'date', 'COUNT'}]
df_covid

```

```python

#df_subway 전처리 (2019-01-01 ~ 2021-07-31)
df_subway = pd.read_csv('./DATA/002_1.DAILY_SUBWAY_INFO.csv', encoding='utf-8', parse_dates=['date'])
# df_subway['date']= df_subway['date'].astype('str')
df_subway['date'] = pd.to_datetime(df_subway['date'].astype('str'))
df_subway = df_subway.drop(['on_num', 'off_num'], axis=1)
df_subway
```

```python
# df_subway, df_covid 합치기
df = pd.merge(df_subway, df_covid, how='outer', on='date')

# <Na>를 0으로 바꾸기
df = df.fillna(0)

# merge 할 때 지맘대로 data type이 float로 변경되서, 다시 int64로 변환
df['COUNT'] = df['COUNT'].astype('Int64')
df['total'] = df['total'].astype('Int64')

df.set_index('date', drop=False, inplace=True)

pd.set_option('display.max_row', 1500)
df

```

```python
# from sklearn.linear_model import LinearRegression
# np.arange(df.date)
np.reshape(-1, 1)

x = df.date
y1 = df.COUNT
y2 = df.total

# Ir = LinearRegression()
# Ir.fit(x, y1, y2)



fig, ax1 = plt.subplots()

ax1.plot(x, y1, '-s', color='green', markersize=7, linewidth=5, alpha=0.7, label='Price')
ax1.set_ylim([datetime.date(2020, 1, 24), datetime.date(2021, 7, 31)]) # x축 범위
ax1.set_xlabel('Period')
ax1.set_ylabel('Count')
ax1.tick_params(axis='both', direction='in')

ax2 = ax1.twinx()
ax2.bar(x, y2, color='deeppink', label='Demand', alpha=0.7, width=0.7)
ax2.set_ylim([datetime.date(2020, 1, 24), datetime.date(2021, 7, 31)]) # x축 범위
ax2.set_ylabel('Total')
ax2.tick_params(axis='y', direction='in')

ax1.set_zorder(ax2.get_zorder() + 10)
ax1.patch.set_visible(False)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

plt.show()
```
